```python
import pandas as pd

# Load the data
file_path = 'kayak_data_v02.csv'
df = pd.read_csv(file_path)

# Define NYC boundaries
lat_min, lat_max = 40.4774, 40.9176
lon_min, lon_max = -74.2591, -73.7004

# Filter data
df_filtered = df[(df['latitude'] >= lat_min) & (df['latitude'] <= lat_max) &
                 (df['longitude'] >= lon_min) & (df['longitude'] <= lon_max)]
```

```python
# Save to new CSV file
df_filtered.to_csv('kayak_data_filteredcoordinates.csv', index=False)
```
