```python
# Import dependencies
import pandas as pd
import sqlite3
from sqlalchemy import create_engine
```

```python
restaurant_df=pd.read_csv('./data/kayak_data_v04-2.csv')

restaurant_df.head()
```

```python
restaurant_df=restaurant_df.reset_index()
restaurant_df=restaurant_df.reset_index(drop=True).rename(columns={'index':'restaurant_id'})
restaurant_df.head()
```

```python
connection=sqlite3.connect('../Web Development and Data Presentation/data/kayak_data.sqlite')
```

```python
restaurant_df.to_sql('restaurants',connection,if_exists='replace',index=False, dtype={'restaurant_id': 'INTEGER PRIMARY KEY AUTOINCREMENT'})
```

```python
connection.close()
```
