```python
# Dependencies
import requests
import json
from pathlib import Path
import pandas as pd
from geopy.geocoders import Nominatim
from geopy.geocoders import GoogleV3
import time
```

```python
data = Path("../mike_test/other_data/kayak_data_v02.csv")
```

```python
# Assuming you've already defined the 'data' variable with the path to your CSV file
# Load the CSV file into a Pandas DataFrame
df = pd.read_csv(data)

# Display the top 5 rows of the DataFrame
df.head(5)
```

```python
# Display the list of column titles
column_titles = df.columns
print(column_titles)
```

```python
#Define a function to geocode the missing coordinates using Google's API:
# Function to get coordinates using Google API
def get_coordinates(address):
    # Initialize the geocoder
    geolocator = GoogleV3(api_key="#######")

    try:
        # Geocode the address
        location = geolocator.geocode(address)
        if location:
            return location.latitude, location.longitude
        else:
            return None, None  # Return None if geocoding fails
    except Exception as e:
        print(f"Geocoding failed for {address}: {str(e)}")
        return None, None

# Iterate through the rows and update missing coordinates
for index, row in df.iterrows():
    if pd.isna(row['latitude']) or pd.isna(row['longitude']):
        name_or_address = row['name']  # You can use the restaurant's name or address here
        latitude, longitude = get_coordinates(name_or_address)
        if latitude is not None and longitude is not None:
            # Update the DataFrame with the obtained coordinates
            df.at[index, 'latitude'] = latitude
            df.at[index, 'longitude'] = longitude
            print(f"Updated coordinates for {name_or_address}: ({latitude}, {longitude})")
        time.sleep(1)  # Sleep for a second to avoid overloading the API
```

```python
##########################
import requests

# Define the base URL for Nominatim
base_url = 'https://nominatim.openstreetmap.org/search'

# Loop through the DataFrame and geocode each restaurant
for index, row in df.iterrows():
    name = row['name']
    locality = row['locality']
    
    # Create the query string
    query = f'{name}, {locality}'
    
    # Prepare the request parameters
    params = {
        'q': query,
        'format': 'json',
    }
    
    try:
        # Send a GET request to Nominatim
        response = requests.get(base_url, params=params)
        
        # Parse the JSON response
        data = response.json()
        
        # Check if a location was found
        if data:
            latitude = data[0]['lat']
            longitude = data[0]['lon']
            df.at[index, 'latitude'] = latitude
            df.at[index, 'longitude'] = longitude
        else:
            print(f"No location found for {query}")
    except Exception as e:
        print(f"Error geocoding {query}: {str(e)}")

# Display the first 10 rows of the DataFrame
print(df.head(10))


```

```python
##########################################
import requests

# List of restaurant names and locations
restaurants = [
    "Le Midi Bar & Restaurant, Manhattan, New York",
    "Malai Marke, East Village, New York",
    "The Flatiron Room - Nomad, Midtown, New York",
    "Bocca Di Bacco (Chelsea - 20th St.), Chelsea, New York",
    "Banc Cafe, Midtown, New York"
]

# Initialize an empty list to store the coordinates
coordinates = []

for restaurant in restaurants:
    # Prepare the request URL
    base_url = 'https://nominatim.openstreetmap.org/search'
    params = {'q': restaurant, 'format': 'json'}
    
    # Make the request
    response = requests.get(base_url, params=params)
    
    # Parse the JSON response
    data = response.json()
    
    # Extract the coordinates if the request was successful
    if data:
        lat = data[0]['lat']
        lon = data[0]['lon']
        coordinates.append((lat, lon))
    else:
        coordinates.append(None)  # Append None for failed requests

# Print the coordinates
for i, coordinate in enumerate(coordinates):
    print(f"Restaurant {i+1}: {coordinate}")
```
