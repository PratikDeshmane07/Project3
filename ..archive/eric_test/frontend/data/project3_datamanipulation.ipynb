```python
import pandas as pd
import re

# Load the data
file_path = 'kayak_data_v04_FINALUPLOAD.csv'
df = pd.read_csv(file_path)
```

```python
df
```

```python
# Exclude specific columns
columns_to_exclude = ['reviews','name', 'noise', 'review', 'latitude', 'longitude', 'url']
columns_to_check = [col for col in df.columns if col not in columns_to_exclude]

# Create a DataFrame to store unique values
unique_values_df = pd.DataFrame(columns=columns_to_check)

# Loop through each remaining column and store unique values by row
for column in columns_to_check:
    unique_values = df[column].unique()
    unique_values_df = unique_values_df.append(pd.Series(unique_values, name=column))

# Save the unique values DataFrame to a CSV file
unique_values_df.to_csv('unique_values.csv', index=False)
```

```python
# Define NYC boundaries
lat_min, lat_max = 40.4774, 40.9176
lon_min, lon_max = -74.2591, -73.7004

# Filter data
df_filtered = df[(df['latitude'] >= lat_min) & (df['latitude'] <= lat_max) &
                 (df['longitude'] >= lon_min) & (df['longitude'] <= lon_max)]
```

```python
df['price_per_person'] = df['price_per_person'].str.replace('\$30 and under', '$')
df['price_per_person'] = df['price_per_person'].apply(lambda x: re.sub(r'\$\d+\s?to\s?\$\d+', '$$', x))
df['price_per_person'] = df['price_per_person'].str.replace('\$50 and over', '$$$')
df.head()
```

```python
# Save to new CSV file
df.to_csv('kayak_data_dollarsigns.csv', index=False)
```
